```python
# Programa que permite calcular la ecuación diferencial ordinaria de 
# segundo orden para el movimiento de un resorte
# Autor: Emmanuel Paz Córdova.
# Fecha: Septiembre 27, 2021

import matplotlib.pyplot as plt
import numpy as np
import math

def x1(k, m, t):
  return np.sin(t*np.sqrt(k/m))

def x2(k, m, t):
  return (x1(k, m, t+h)-x1(k, m, t))/h

def x3(k, m, t):
  return (x2(k, m, t+h)-x2(k, m, t))/h

t1=np.arange(0.00, 3.00, 0.01) 

print("Ingrese la constante del resorte (en N/m): ")
constante=float(input())
print("Ingrese la masa (en kg): ")  
masa=float(input())
print("Ingrese el tiempo (en s): ")
tiempo=float(input())
print("Ingrese el valor del incremento")
h=float(input())

print("El valor del desplazamiento en el tiempo dado es: ", x1(constante, masa, tiempo))
print("El valor de la velocidad en el tiempo dado es: ", x2(constante, masa, tiempo))
print("El valor de la aceleración en el tiempo dado es: ", x3(constante, masa, tiempo))

# Gráfica de la función original
plt.plot(t1, x1(constante, masa, t1), label='x(t)')
# Gráfica de la primera derivada de la función (velocidad)
plt.plot(t1, x2(constante, masa, t1), label='x´(t)')
# Gráfica de la segunda derivada de la función (aceleración)
plt.plot(t1, x3(constante, masa, t1), label='x´´(t)')
# Nombre de los ejes x y y
plt.xlabel("Variable Independiente (t) [units]")
plt.ylabel("Variable Dependiente f(t) [units]")
# Impresión del nombre de cada función
plt.legend()
# Creación de la gráfica
plt.show()
```
